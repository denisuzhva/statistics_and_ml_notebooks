```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# A given covariance matrix
cov_mat = np.array([[2, 1.5], [1.5, 2]])

# Mean values
means = np.array([0, 0])
```

```python
# Number of samples
n_samp = 10000

sample_points = np.random.multivariate_normal(means, cov_mat, n_samp)
X = sample_points[:, 0]
Y = sample_points[:, 1]
```

```python
plt.figure(figsize=(4, 4), dpi=150)
plt.scatter(X, Y, s=0.3, c='k')
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.show()
```

```python
# Calculate covariance

XY = X*Y
XY_mean = np.mean(XY)
X_mean = np.mean(X)
Y_mean = np.mean(Y)
cov_calc = np.mean(XY_mean - X_mean*Y_mean)
print(cov_calc)
```

```python
# Plot mutually multiplied data points and their average values

plt.figure(figsize=(4, 4), dpi=150)
plt.hist(XY, bins=1000, color='k')
plt.axvline(x=XY_mean, color='green', label='XY mean')
plt.axvline(x=X_mean, color='red', label='X mean')
plt.axvline(x=Y_mean, color='blue', label='y mean')
plt.xlabel('XY', fontsize=12)
plt.xlim(-2, 4)
plt.show()
```

Partial histogram of the mutually multiplyed data points. 
The green line corresponds to the average value of $X \odot Y$, while the red and the blue ones correspond the average values of $X$ and $Y$ respectively. 
The red and the blue ones are close to 0, since $X$ and $Y$ obey the normal distribution with zero mean. 
The higher the elongation of the "cloud" on the $X$-$Y$ plane (along the line $y = f(x) = x$), the more data points of $X$ correspond to data points of $Y$ of the same sign, which leads to the histogram outweighed toward the positive values.
